```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://cdn3.gnarususercontent.com.br/3137-data-visualization/Projeto/Dados/pib_br_2002_2020_estados.csv")
df.sample(5)
```

     Deixamos como desafio a ideia de construir uma visualização para representar os estados de uma região. Que tal testar isto para a região Sul, por exemplo?

```python
df_sul = df.query("regiao == 'Sul'").groupby("sigla_uf")[["pib"]].sum()
```

```python
df_sul
```

```python
fig, ax = plt.subplots(dpi=150)
fig.suptitle("Composição do PIB da Região Sul do País.")
fig.set_facecolor(color="#939393")
wedges, text, autotexts = ax.pie(data=df_sul,x="pib",labels=df_sul.index, startangle=90, autopct="%.1f%%")
for i, t in enumerate(text):
    t.set_color(wedges[i].get_facecolor())
for t in autotexts:
    t.set_color("white")
    t.set_weight("bold")
# ax.tick_params(axis="x", which="major", direction="in")
# ax.tick_params(axis="y", which="major", length=0)
```

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

````python
vendas = pd.read_csv(
    "https://raw.githubusercontent.com/alura-cursos/dataviz-graficos/master/dados/relatorio_vendas.csv"
)
vendas.head()
````

````python
vendas[["data_pedido", "data_envio"]].info()
````

````python
for col in ["data_pedido", "data_envio"]:
    vendas[col] = pd.to_datetime(vendas[col], format="%Y-%m-%d")

vendas[["data_pedido", "data_envio"]].info()
````

````python
vendas_anuais = vendas[["data_pedido", "vendas"]].copy()
vendas_anuais["ano"] = vendas_anuais.data_pedido.dt.year

vendas_anuais = vendas_anuais.drop(columns="data_pedido")
````

````python
# Agrupar por ano
df_vendas_anuais = vendas_anuais.groupby(by="ano").sum()
````

`sns.set_style =! set_theme`

````python
sns.set_theme(style="white")
fig, ax = plt.subplots(figsize=(10, 4), dpi=300, layout="constrained")
fig.suptitle("Vendas das lojas de departamento", x=0.20)


# Plot
ax = sns.barplot(data=df_vendas_anuais, x=df_vendas_anuais.index, y="vendas")

ax.set_ylabel("")
ax.set_xlabel("")


# Ticks
ax.xaxis.set_tick_params(labelsize=14, labelcolor=CINZA2)
ax.set_yticklabels([])

# Add value labels on top of each bar
for bar in ax.patches:
    value = f"{bar.get_height():,.0f}".replace(",", ".")
    x = bar.get_x() + bar.get_width() / 2
    y = bar.get_height()
    ax.annotate(
        f"R$ {value}", (x, y), ha="center", va="bottom", fontsize=12, color=CINZA2
    )
sns.despine(left=True)

# ax.bar_label()
````

````python
# Função que plota o gráfico.
from matplotlib.axes import Axes


def plot_bars_with_pallete(cores: list[str] = [AZUL2]) -> Axes:
    sns.set_theme(style="white")
    fig, ax = plt.subplots(figsize=(10, 4), dpi=300, layout="constrained")
    fig.suptitle("Vendas das lojas de departamento", x=0.20)

    ax = sns.barplot(
        data=df_vendas_anuais, x=df_vendas_anuais.index, y="vendas", palette=cores
    )

    ax.set_ylabel("")
    ax.set_xlabel("")

    # Ticks
    ax.xaxis.set_tick_params(labelsize=14, labelcolor=CINZA2)
    ax.set_yticklabels([])

    # Add value labels on top of each bar
    for bar in ax.patches:
        value = f"{bar.get_height():,.0f}".replace(",", ".")
        x = bar.get_x() + bar.get_width() / 2
        y = bar.get_height()
        ax.annotate(
            f"R$ {value}", (x, y), ha="center", va="bottom", fontsize=12, color=CINZA2
        )
    sns.despine(left=True)

    return ax
````

````python
cores = []

for ano in df_vendas_anuais.index:
    if df_vendas_anuais.loc[ano, "vendas"] == df_vendas_anuais["vendas"].max():
        cores.append(AZUL2)
    else:
        cores.append(AZUL5)

ax = plot_bars_with_pallete(cores=cores)


# Anotando uma conclusão no gráfico
ax.text(
    3.5,
    1.5e6,
    "Em $\\bf{2019}$, as vendas\n"
    "nas lojas subiram\n"
    "aproximadamente $\\bf{22,3}$%\n"
    "em relação ao ano de 2018.",
    fontsize=14,
    linespacing=1.45,
    color=AZUL2,
)
fig = ax.get_figure()
````

Desafio 2: Qual foi o faturamento (vendas) dos top 10 produtos durante o período de 2016 a 2019 do nosso conjunto de dados? Adicione um pequeno texto falando dos 3 produtos que mais venderam.

````python
faturamento = vendas[["tipo_produto", "vendas"]].copy()
````

````python
fat10 = (
    faturamento.groupby("tipo_produto")
    .agg("sum")
    .sort_values(by="vendas", ascending=False)
    .reset_index()
    .loc[:9]
)
````

````python
bar_colors = [AZUL2 if i <= 3 else CINZA2 for i in range(fat10.tipo_produto.size)]


fig, ax = plt.subplots(dpi=600)
fig.subplots_adjust(right=0.7)
ax = sns.barplot(data=fat10, x=fat10.vendas, y=fat10.tipo_produto, palette=bar_colors)
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_xticks([])

for i, valor in enumerate(fat10["vendas"]):
    qtd = f"R$ {valor:,.0f}".replace(",", ".")
    offset = 1e4  # offset de 10.000
    ax.text(
        valor - offset,
        i,
        qtd,
        color=CINZA5,
        fontsize=10,
        fontweight="bold",
        ha="right",
        va="center",
    )

sns.despine(left=True, bottom=True)
plt.show()
````

> Qual o modo de envio mais utilizado pelos clientes da loja? É proporcional para B2B e B2C?

````python
cols_interesse = ["modo_envio", "segmento_cliente"]
df_envio = vendas[cols_interesse].copy()

ct_envio_segmento = pd.crosstab(
    index=df_envio["modo_envio"],
    columns=df_envio["segmento_cliente"]
).sort_values(by="B2B")

ct_envio_segmento
````

````python
ct_envio_segmento_norm = pd.crosstab(
    index=df_envio["modo_envio"],
    columns=df_envio["segmento_cliente"],
    normalize="index" # %  
).sort_values(by="B2B")

ct_envio_segmento_norm
````

````python
ct_envio_segmento
````

````python

print("="*80)
print("B2B\n",ct_envio_segmento["B2B"] / ct_envio_segmento.sum(axis=1)) # soma por modo de envio
print("="*80)
print("B2C\n",ct_envio_segmento["B2C"] / ct_envio_segmento.sum(axis=1)) # soma por modo de envio
print("="*80)
````

````python
CORES = [AZUL2, AZUL5]
TITLE = "Modos de envio mais utilizados pelos clientes 2016-2019"

fig, ax = plt.subplots(
    1, 2, figsize=(14, 6), dpi=300, layout="constrained", sharey=True
)
fig.suptitle(t=TITLE, fontweight="bold", x=0.2)

# ax[0].set_title("Valores Absolutos")
ax[1].set_title("50 (%) do Total de Vendas", color="#e81426")
sns.despine(bottom=True, ax=ax[0])
sns.despine(left=True, bottom=True, ax=ax[1])
for axi in ax:
    axi.set_xticks([])


for i, df in enumerate([ct_envio_segmento, ct_envio_segmento_norm]):
    ax[i].barh(df.index, df["B2B"], color = cores[0], label=df.T.index[0])
    ax[i].barh(df.index, df["B2C"], left = df["B2B"], color = cores[1], label=df.T.index[1])
    

# Valores das barras
for container in ax[0].containers:
    labels = [str(valor.get_width()) if valor.get_width() > 300 else '' for valor in container]
    ax[0].bar_label(container, label_type='center', labels = labels, size = 12, color = CINZA5, fontweight='bold')

for container in ax[1].containers:
    labels = [f'{valor.get_width()*100:.1f}%' for valor in container]
    ax[1].bar_label(container, label_type='center', labels = labels, size = 12, color = CINZA5, fontweight='bold')
ax[1].axvline(0.5, color="#e81426", ls=':')

ax[0].legend(frameon=True, loc="upper center", bbox_to_anchor=(1.0, 1.1), fontsize=12, reverse=False, ncols=3)
plt.show()

````

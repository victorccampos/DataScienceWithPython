```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imigrantes_canada.csv')
df
```

<h2 align="center">An√°lise de tend√™ncias de migra√ß√£o do Brasil em um determinado per√≠odo </h2> 

```python
# Tornar o index do dataframe como o pa√≠s
df.set_index('Pa√≠s', inplace=True)
df.head(10)
```

```python
# Armarzenar o intervalo de tempo
anos: list[str] = list(map(str, range(1980,2014)))
anos
```

```python
# Ano x N¬∫ de Imigrantes
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {
    'ano' : brasil.index.tolist(),
    'imigrantes' : brasil.values.tolist()
}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) 
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title("Imigra√ß√£o do Brasil para o Canad√°")
# R√≥tulos
plt.xlabel('Ano')
plt.ylabel('N√∫mero de Imigrantes')
plt.show()
```

```python
# Desafio: criar um gr√°fico de linhas comparando os n√∫meros de imigrantes do Brasil e Argentina para o Canad√°
plt.figure(figsize=(8,4))
plt.plot(df.loc['Argentina', anos].index.tolist() , df.loc['Argentina', anos].values.tolist(), label='Argentina')
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title("Compara√ß√£o de Imigra√ß√£o: Brasil e Argentina para o Canad√°")
# R√≥tulos
plt.xlabel('Ano')
plt.ylabel('N√∫mero de Imigrantes')
plt.legend()
plt.show()
```

<h3>Trabalhando com Figuras em Matplotlib - (Abordagem Orientada a Objetos) </h3>

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.set_title("Imigra√ß√£o do Brasil para o Canad√°\n1980-2013")
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # Frequ√™ncia de 5 em 5 no eixo x
ax.set_xlabel('Ano')
ax.set_ylabel('N¬∫ de Imigrantes')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(
    nrows=1,
    ncols=2,
    figsize=(15,5)
)
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigra√ß√£o do Brasil para o Canad√°\n1980-2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('N¬∞ de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))


axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigra√ß√£o do Brasil para o Canad√°\n1980-2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('N¬∞ de Imigrantes')
```

```python
fig, axs = plt.subplots(
    nrows=2, 
    ncols=2,
    figsize = (10, 6)
)
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Col√¥mbia', anos])
axs[0,1].set_title('Col√¥mbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')


axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Ano')
    ax.set_ylabel('N¬∫ de Imigrantes')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
ax.barh(america_do_sul.index, america_do_sul['Total'], color=cores)
ax.set_title('Imigra√ß√£o da Am√©rica do Sul para o Canad√°\n1980-2013', fontsize=18)
ax.set_xlabel("N¬∫ de Imigrantes", fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
america_sul_sorted = america_do_sul.sort_values('Total', ascending=True) # Maior pro menor
fig, ax = plt.subplots(figsize=(12,6))

cores = ['green'  if pais == 'Brasil' else 'silver' for pais in america_sul_sorted.index]

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Am√©rica do Sul: Brasil foi o quarto pa√≠s com mais imigrante para o Canad√°\n1980-2013', fontsize=18, loc='left')
ax.set_xlabel("N¬∫ de Imigrantes", fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Valores maximo de cada barra 
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(x=v+20, y=i, s=str(v), color='black', fontsize=12, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)
plt.show()
```

https://seaborn.pydata.org/tutorial/color_palettes.html : Paletas de cores do Seaborn

```python
def gerar_grafico_paleta(paleta):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top10, y=top10.index, x='Total', orient='h', hue='Total', palette=paleta, legend=False)
    ax.set_title('Pa√≠ses com maior taxa de imigra√ß√£o para o Canad√°\n1980-2013',loc='left', fontsize=16)
    ax.set_xlabel('N¬∫ de imigrantes', fontsize=14)
    ax.set_ylabel('', fontsize=14)
    ax.grid(color='silver', linestyle='--')    
    sns.despine()
    plt.show()
    
```

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')

```

- Desafio  
Nesta etapa, seu desafio √© criar uma figura contendo as tend√™ncias de imigra√ß√£o dos 4 maiores pa√≠ses da Am√©rica latina: Brasil, Argentina, Peru e Col√¥mbia. Atrav√©s dessa cria√ß√£o voc√™ pode explorar diversas possibilidades e reconhecer de forma atrativa o seu processo de desenvolvimento.E n√£o nos esque√ßamos das orienta√ß√µes! Essa figura precisa ter uma linha para cada pa√≠s, t√≠tulo, r√≥tulos nos eixos, cores apropriadas, um tema da biblioteca Seaborn e legenda. Por isso, pense nas quest√µes de acessibilidade, como tamanho das fontes e espessura das linhas.

```python
sns.set_theme()
sns.set_palette('Dark2')

fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.lineplot(df.loc['Brasil', anos], label='Brasil', lw=3)
ax = sns.lineplot(df.loc['Argentina', anos], label='Argentina', lw=3)
ax = sns.lineplot(df.loc['Peru', anos], label='Peru', lw=3)
ax = sns.lineplot(df.loc['Col√¥mbia', anos], label='Col√¥mbia', lw=3)

ax.set_title('Imigra√ß√£o dos maiores pa√≠ses da Am√©rica do Sul\npara o Canad√° de 1980 a 2013', loc='left', fontsize=20)
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('N√∫mero de imigrantes', fontsize=14)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

ax.legend(title='Pa√≠ses', loc='upper right', bbox_to_anchor=(1.18, 1.02))

plt.show()
```

<h3> 📊 O que é "Pivot" em análise de dados? </h3>

> __"Pivot" significa transformar dados de formato "longo" para formato "largo", reorganizando linhas em colunas.__

<div style="display: flex; gap: 40px;">

  <div>

  🔹 <strong>Exemplo simples — antes do "pivot"</strong>  
  <table>
    <thead>
      <tr><th>Pessoa</th><th>Ano</th><th>Valor</th></tr>
    </thead>
    <tbody>
      <tr><td>Ana</td><td>2020</td><td>100</td></tr>
      <tr><td>Ana</td><td>2021</td><td>150</td></tr>
      <tr><td>João</td><td>2020</td><td>200</td></tr>
      <tr><td>João</td><td>2021</td><td>250</td></tr>
    </tbody>
  </table>
  <p>Esse é o formato longo ("long format"): cada linha é uma entrada com múltiplas variáveis.</p>

  </div>

  <div>

  🔸 <strong>Após "pivot"</strong>  
  <table>
    <thead>
      <tr><th>Pessoa</th><th>2020</th><th>2021</th></tr>
    </thead>
    <tbody>
      <tr><td>Ana</td><td>100</td><td>150</td></tr>
      <tr><td>João</td><td>200</td><td>250</td></tr>
    </tbody>
  </table>
  <p>Agora está em formato largo ("wide format"). Isso é útil para análises que comparam colunas, gráficos, matrizes, etc.</p>

  </div>

</div>

````python
cols_info = list(emissoes_gases.columns[:10])
cols_emissao_values = list(emissoes_gases.columns[10:])

emissoes_por_ano = emissoes_gases.melt(
    id_vars = cols_info,
    value_vars = cols_emissao_values,
    var_name = 'Ano',
    value_name = 'Emissão'
)
````

Trocamos então 52 colunas de ano, onde cada linha era uma emissão por categoria, por duas colunas `Ano` e `Emissão`, contendo a mesma informação.

````python
emissoes_gases
````

atributo `groups` e seleções específicas com método `get_group()`

````python
emissoes_por_ano.groupby('Gás').groups
````

````python
emissoes_por_ano.groupby('Gás').groups.keys()
````

````python
emissoes_por_ano.groupby('Gás').get_group('CO2 (t)') # emissão de CO2  
````

Os métodos de agregação que podemos utilizar com o groupby() são:

- count() para realizar contagem;  
- sum() para soma;  
- mean() para a média;  
- median() para a mediana;  
- min() para o valor mínimo;  
- max() para o valor máximo;  
- std() para o desvio-padrão;  
- var() para a variância.  

__Soma da emissão total por tipo de gás__

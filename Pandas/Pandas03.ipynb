<h3> üìä O que √© "Pivot" em an√°lise de dados? </h3>

> __"Pivot" significa transformar dados de formato "longo" para formato "largo", reorganizando linhas em colunas.__

<div style="display: flex; gap: 40px;">

  <div>

  üîπ <strong>Exemplo simples ‚Äî antes do "pivot"</strong>  
  <table>
    <thead>
      <tr><th>Pessoa</th><th>Ano</th><th>Valor</th></tr>
    </thead>
    <tbody>
      <tr><td>Ana</td><td>2020</td><td>100</td></tr>
      <tr><td>Ana</td><td>2021</td><td>150</td></tr>
      <tr><td>Jo√£o</td><td>2020</td><td>200</td></tr>
      <tr><td>Jo√£o</td><td>2021</td><td>250</td></tr>
    </tbody>
  </table>
  <p>Esse √© o formato longo ("long format"): cada linha √© uma entrada com m√∫ltiplas vari√°veis.</p>

  </div>

  <div>

  üî∏ <strong>Ap√≥s "pivot"</strong>  
  <table>
    <thead>
      <tr><th>Pessoa</th><th>2020</th><th>2021</th></tr>
    </thead>
    <tbody>
      <tr><td>Ana</td><td>100</td><td>150</td></tr>
      <tr><td>Jo√£o</td><td>200</td><td>250</td></tr>
    </tbody>
  </table>
  <p>Agora est√° em formato largo ("wide format"). Isso √© √∫til para an√°lises que comparam colunas, gr√°ficos, matrizes, etc.</p>

  </div>

</div>

````python
cols_info = list(emissoes_gases.columns[:10])
cols_emissao_values = list(emissoes_gases.columns[10:])

emissoes_por_ano = emissoes_gases.melt(
    id_vars = cols_info,
    value_vars = cols_emissao_values,
    var_name = 'Ano',
    value_name = 'Emiss√£o'
)
````

````python
# pip install xlrd
pop_estados = pd.read_excel('data/POP2022_Municipios.xls', 
              header = 1,
              skipfooter = 34)
pop_estados
````

````python
pop_estados.info()

# Coluna POPULA√á√ÉO est√° em String
````

````python
# Busca desse tipo de erro na coluna utilizando regex

re_pattern : str = r'\('

pop_estados[
    pop_estados['POPULA√á√ÉO'].str.contains(pat=re_pattern, na=False)  # Boolean Array
]
````

````python
### Remo√ß√£o de Par√™nteses e '.' 

# Novas colunas a partir de cols j√° existentes
padrao_regex = r'\(\d{1,2}\)'

pop_estados = pop_estados.assign(
    pop_sem_parenteses = pop_estados['POPULA√á√ÉO'].replace(padrao_regex, '', regex=True),
    populacao = lambda x: x.loc[:,'pop_sem_parenteses'].replace('\.', '', regex=True)
    )
````

````python
re_pattern : str = r'\('

pop_estados[
    pop_estados['POPULA√á√ÉO'].str.contains(pat=re_pattern, na=False)  # Boolean Array
]
````

````python
# Agora finalmente pode transformar para valor num√©rico
pop_estados = pop_estados.astype(
    {'populacao': 'int64'}
)
print(pop_estados.info())
````

````python
pop_estados
````

````python
pop_estados = pop_estados.groupby('UF')[['populacao']].sum()
pop_estados
````

````python
pop_estados = pop_estados.reset_index() # Transforma novamente o Index em 1,2,3,4,...
````

````python
pop_estados
````

__Merge__ de tabelas

Primeiro, vamos gerar uma nova tabela com o agrupamento da emiss√£o para cada estado somente do ano mais recente que temos no dataframe, 2021. Depois, vamos unir uma tabela √† outra: emissao_estados e populacao_estados.

````python
emissao_estados_2021 = emissoes_por_ano[
    emissoes_por_ano['Ano'] == 2021
    ].groupby('Estado')[['Emiss√£o']].sum().reset_index()
emissao_estados_2021
````

````python
emissao_per_capita = pd.merge(left=emissao_estados_2021, right=pop_estados, left_on='Estado', right_on='UF')
emissao_per_capita
````

````python
emissao_per_capita.plot(kind='scatter', x='populacao', y='Emiss√£o');
````

3. `pd.merge()`.

O m√©todo `assign()` permite criar uma ou mais colunas com base em opera√ß√µes com colunas existentes e retorna uma c√≥pia do DataFrame original com as novas colunas adicionadas. Para criar colunas com base em uma coluna rec√©m-criada, √© necess√°rio utilizar uma fun√ß√£o lambda na cria√ß√£o da coluna.

````python
emissao_per_capita = emissao_per_capita.assign(
    per_capita = emissao_per_capita['Emiss√£o'] / emissao_per_capita['populacao']
).sort_values(by='per_capita', ascending=False)

emissao_per_capita
````

__Tipos inteiros__

```python
# max_hospedes, quantidade_*, 
cols_int = ['max_hospedes'] + [col for col in dados.columns if col.startswith('quantidade')]
cols_int
```

```python
dados[cols_int] = dados[cols_int].astype(np.int64)
```

```python
dados.info()
```

__Tipos float__

```python
# Formatacao 
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',','').strip())
# Tipo
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
# Formatação
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].map(lambda x: x.replace('$', '').replace(',','').strip())
# Tipo
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

```python
dados['avaliacao_geral'].unique() # pode ser float
```

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

## <font color='orange'> Dados textuais </font>  

Foco na coluna `descricao_local`: tokenização.

```python
dados['descricao_local'][3169]
```

método `replace()`

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
regex_pat: str = r"[^a-zA-Z0-9\-\']"
dados['descricao_local'].str.replace(pat=regex_pat, repl=' ', regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace(pat=regex_pat, repl=' ', regex=True)
```

```python
# Remover hífen de palavras que não são compostas.
# don't look behind
re_hifen_pat : str = r"(?<!\w)-(?!\w)"   
dados['descricao_local'] = dados['descricao_local'].str.replace(pat=re_hifen_pat, repl=' ', regex=True)
```

__Tokenização__


```python
# separação de palavras em listas => split()
barra_de_espaço: str = " "
dados['descricao_local'] = dados['descricao_local'].str.split(barra_de_espaço)
```

```python
dados.head()
```

__Comodidades__

```python
re_chave: str = r'\{|}|\"'
dados['comodidades'] = dados['comodidades'].str.replace(pat=re_chave, repl=" ", regex=True)
```

```python
dados.head(2)
```

__Desafio__: Tokenização coluna `descricao_vizinhanca`

```python
# dados['descricao_vizinhanca'][402]

dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace(pat=regex_pat, repl=' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace(pat=re_hifen_pat, repl=' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split(" ") # List[str]
```

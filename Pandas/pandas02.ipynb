No exemplo acima, verifiquei que não é necessário usar o BeautifulSoup. Segue a justificativa do porquê abaixo:  

O **pandas** consegue lidar diretamente com URLs porque internamente ele usa a biblioteca **lxml** (ou **html5lib**, dependendo do parser disponível) para buscar e processar o HTML. Então, quando você passa uma URL para `pd.read_html()`, ele já faz o download do conteúdo, parseia e extrai as tabelas automaticamente—por isso deu conta sem precisar do **requests**.

Agora, por que você pensou que precisaria do **requests**? Isso pode ter vindo da ideia de que, antes de processar algo, precisamos primeiro _pegá-lo_. Muitas bibliotecas de análise de dados ou scraping exigem que você manualmente busque o conteúdo da web antes de processá-lo. Então, instintivamente, parece lógico pensar: "preciso pegar o HTML primeiro e depois passar para o pandas". Mas, no caso do `pd.read_html()`, ele faz isso por baixo dos panos.

 __Quando usar `requests`?__
Você deve usar **requests** quando:
1. **Precisa manipular o HTML antes de passá-lo para outra biblioteca** – Por exemplo, remover partes indesejadas, modificar o encoding ou reformatar o texto.
2. **A página requer autenticação ou headers especiais** – Alguns sites bloqueiam acesso direto a bibliotecas como pandas, então você pode precisar configurar cookies, headers ou autenticação.
3. **Quer armazenar o HTML para uso posterior** – Se precisar salvar o conteúdo para analisar offline ou comparar versões.
4. **Precisa acessar APIs** – `requests` é essencial para obter dados de APIs JSON e outros endpoints web.

Em resumo, `pd.read_html()` funciona bem quando você quer extrair tabelas rapidamente de uma página, sem complicações. Mas **requests** te dá mais controle quando precisa lidar com conteúdos de forma mais customizada.
